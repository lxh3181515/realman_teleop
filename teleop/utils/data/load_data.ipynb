```python
import numpy as np
import matplotlib.pyplot as plt

# 加载数据
head_record_path = 'head_record.npy'
head_record = np.load(head_record_path, allow_pickle=True)

# 获取时间维度长度和维度数量
timesteps, num_dims = head_record.shape

# 创建画布和子图
plt.figure(figsize=(15, 10))

for i in range(num_dims):
    plt.subplot(num_dims, 1, i + 1)
    plt.plot(range(timesteps), head_record[:, i], label=f'Dimension {i+1}')
    plt.ylabel(f'Dim {i+1}')
    plt.grid(True)
    if i == 0:
        plt.title('Visualization of 6 Dimensions over Time')
    if i == num_dims - 1:
        plt.xlabel('Timestep')
    plt.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
